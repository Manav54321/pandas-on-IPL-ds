```python
data2 = pd.read_csv('deliveries.csv')
data2.head()
```

```python
data2.shape
```

```python
data2.info()
```

```python
# find top 10 batsman in terms of runs
batsman = data2.groupby('batsman')
batsman['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with maximum no. of sixes
data2['batsman_runs'] == 6
sixes = data2[data2['batsman_runs'] == 6]
max_six = sixes.groupby('batsman')
max_six['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python
# find the batsman with max no. of 4 and 6 in the last 5 overs
over15 = data2['over'] > 15
data2[over15].head()
run4 = data2['batsman_runs'] == 4
run6 = data2['batsman_runs'] == 6
max_46 = data2[over15 & (run4 | run6)]
max_46.groupby('batsman').count().sort_values('ball',ascending=False).head(1).index[0]
```

```python
# find MS Dhoni record against all team
data2['batsman'] == 'MS Dhoni'
ms = data2[data2['batsman'] == 'MS Dhoni']
ms.groupby('bowling_team').count().sort_values('ball',ascending=False).head(20)['batsman_runs']
# proved because there is no CSK or RPSG team occured in the o/p, MS D in those team
```

```python
# create a function of a batsman who scored max. run in all of the seasons.

def max_runs_batsman(data):
  batsman = data.groupby('batsman')
  return batsman['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
max_runs_batsman(data2)
```

# Merge in Pandas, same as Join in SQL.

```python
# orange cap holder in every seasons.
new = pd.merge(data, data2, left_on='id', right_on='match_id')
new.head()
```
